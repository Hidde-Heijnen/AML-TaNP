```python
# Import necessary libraries
import os
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to load dataset
def load_dataset(dataset_path):
    """Load user_list, item_list, and interaction_dict_x from a dataset directory."""
    with open(os.path.join(dataset_path, 'user_list.json'), 'r') as f:
        user_list = json.load(f)
    with open(os.path.join(dataset_path, 'item_list.json'), 'r') as f:
        item_list = json.load(f)
    with open(os.path.join(dataset_path, 'interaction_dict_x.json'), 'r') as f:
        inter_dict_x = json.load(f)
    return user_list, item_list, inter_dict_x
```

```python

# Function to analyze dataset
def analyze_dataset(user_list, item_list, inter_dict_x):
    """Calculate statistics for the dataset."""
    num_users = len(user_list)
    num_items = len(item_list)
    num_interactions = sum(len(interactions) for interactions in inter_dict_x.values())
    interactions_per_user = [len(interactions) for interactions in inter_dict_x.values()]
    return num_users, num_items, num_interactions, interactions_per_user
```

```python

# Function to plot interaction distribution
def plot_interactions(interactions_per_user, title):
    """Plot histogram of interactions per user."""
    plt.hist(interactions_per_user, bins=50, edgecolor='black')
    plt.title(title)
    plt.xlabel('Number of Interactions')
    plt.ylabel('Number of Users')
    plt.show()
```

```python

# Define dataset paths (update these to your actual paths)
lastfm_20_path = './TaNP/data/lastfm_20'
lastfm_hetrec_path = './TaNP/data/lastfm_hetrec'

# Load datasets
lastfm_20_users, lastfm_20_items, lastfm_20_inter_x = load_dataset(lastfm_20_path)
lastfm_hetrec_users, lastfm_hetrec_items, lastfm_hetrec_inter_x = load_dataset(lastfm_hetrec_path)

# Analyze datasets
l20_users, l20_items, l20_interactions, l20_inter_per_user = analyze_dataset(
    lastfm_20_users, lastfm_20_items, lastfm_20_inter_x
)
lhet_users, lhet_items, lhet_interactions, lhet_inter_per_user = analyze_dataset(
    lastfm_hetrec_users, lastfm_hetrec_items, lastfm_hetrec_inter_x
)

```

```python
# Print statistics
print("### lastfm_20 Statistics ###")
print(f"Number of users: {l20_users}")
print(f"Number of items (artists): {l20_items}")
print(f"Number of interactions: {l20_interactions}")
print(f"Average interactions per user: {np.mean(l20_inter_per_user):.2f}")

print("\n### lastfm_hetrec Statistics ###")
print(f"Number of users: {lhet_users}")
print(f"Number of items (artists): {lhet_items}")
print(f"Number of interactions: {lhet_interactions}")
print(f"Average interactions per user: {np.mean(lhet_inter_per_user):.2f}")

```

```python
# Plot interaction distributions
plot_interactions(l20_inter_per_user, 'lastfm_20: Interactions per User')
plot_interactions(lhet_inter_per_user, 'lastfm_hetrec: Interactions per User')
```
